```python
import nltk
from nltk.tokenize import sent_tokenize
from nltk.tokenize import word_tokenize
from nltk.stem.porter import PorterStemmer
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
import re
```

```python
def sentence_tokenization(s):
    sentences_list = sent_tokenize(s)
    
    return sentences_list

def removeNoise(str):
    str =  re.sub(r"\s", " ", str)
    new_string =  re.sub(r"\([\w.]*\)", " ", str)
    return new_string

def casefolding(s):
    new_str = s.lower()
    
    return new_str

def lemmatize(lem):
    wordnet_lemmatizer = WordNetLemmatizer()
    result = list()
    
    nltk_tokens = nltk.word_tokenize(lem)
    for w in nltk_tokens:
        result.append(wordnet_lemmatizer.lemmatize(w))
    return ' '.join(result)

def mainEng():
    f = open ('jsbacheng.txt','r')
    text_data = f.read()
    f.close()
    print(text_data)
    text_data = removeNoise(text_data)
    text_data = casefolding(text_data)
    text_data = lemmatize(text_data)
    print(sentence_tokenization(text_data))
    
if __name__ == "__main__":
    mainEng()
    
```
